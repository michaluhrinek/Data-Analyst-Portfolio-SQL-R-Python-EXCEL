```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file into a DataFrame
file_path = r"C:\Users\User\Downloads\flooding_data_sk.xlsx"
df = pd.read_excel(file_path)

# Convert the "Date" column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Group by date and count the number of events on each date
events_over_time = df.groupby('Date').size()

# Create a scatterplot
plt.figure(figsize=(10, 6))

# Scatter plot
plt.scatter(events_over_time.index, events_over_time.values, color='blue', marker='o')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Number of Flooding Events')
plt.title('Flooding Events Over Time (Scatterplot)')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
df_2023 = df[df['Date'].dt.year == 2023]

# Group by date and count the number of events on each date for 2023
events_over_time_2023 = df_2023.groupby('Date').size()

# Create a scatterplot
plt.figure(figsize=(10, 6))

# Scatter plot
plt.scatter(events_over_time_2023.index, events_over_time_2023.values, color='blue', marker='o')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Number of Flooding Events')
plt.title('Flooding Events in 2023')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python

# Extract the year from the "Date" column
df['Year'] = df['Date'].dt.year

# Group by year and sum the cost of damage for each year
cost_of_damage_by_year = df.groupby('Year')['cost of damage (milion)'].sum()

# Create a bar chart
plt.figure(figsize=(10, 6))

# Bar plot
bars = plt.bar(cost_of_damage_by_year.index, cost_of_damage_by_year.values, color='red')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Cost of Damage (Million)')
plt.title('Cost of Damage from Flooding Events by Year')

# Adding cost of damage values on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom', ha='center', color='black', fontsize=9)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
#Group by location and count the number of events for each river
events_location = df['Location'].value_counts()
#Create a bar chart
plt.figure(figsize=(10,6))
#bar plot
bars = plt.bar(events_location.index, events_location.values, color='Red')

#adding labels
plt.xlabel('Location')
plt.ylabel('Number of Flooding Events')
plt.title('Number of Flooding events by Location')

#rotation x-axis labels for better readability
plt.xticks(rotation=45)

#Adding event counts on top of each bar
for bar in bars: 
    yval = bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2, yval, int(yval), va='bottom', ha='center', color='black',fontsize=9)
#Display the plot
plt.tight_layout()
plt.show()
```

```python
# Group by river and count the number of events for each river
events_by_river = df['River'].value_counts()

# Create a bar chart
plt.figure(figsize=(10, 6))

# Bar plot
bars = plt.bar(events_by_river.index, events_by_river.values, color='blue')

# Adding labels and title
plt.xlabel('River ')
plt.ylabel('Number of Flooding Events')
plt.title('Number of Flooding Events by River')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Adding event counts on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center', color='black', fontsize=9)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Group by location and sum the total number of people affected by flooding in each location
people_affected_by_location = df.groupby('Location')['people affected by flooding'].sum()

# Create a bar chart
plt.figure(figsize=(12, 6))

# Bar plot
bars = plt.bar(people_affected_by_location.index, people_affected_by_location.values, color='blue')

# Adding labels and title
plt.xlabel('Location')
plt.ylabel('Total People Affected by Flooding')
plt.title('Total People Affected by Flooding in Each Location')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adding total people affected on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center', color='black', fontsize=9)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Group by location and sum the total number of people affected by flooding in each location
people_affected_by_location = df.groupby('Location')['people affected by flooding'].sum()

# Sort the data in descending order of total people affected
people_affected_by_location = people_affected_by_location.sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(12, 6))

# Bar plot
bars = plt.bar(people_affected_by_location.index, people_affected_by_location.values, color='blue')

# Adding labels and title
plt.xlabel('Location')
plt.ylabel('Total People Affected by Flooding')
plt.title('Total People Affected by Flooding in Each Location (Descending Order)')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adding total people affected on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center', color='black', fontsize=9)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Create a line chart
plt.figure(figsize=(10, 6))

# Line plot for rainfall over time
plt.plot(df['Date'], df['rainfall (mm)'], marker='o', linestyle='-', color='blue')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Rainfall (mm)')
plt.title('Rainfall Levels Over Time for Each Flooding Event')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Create a line chart
plt.figure(figsize=(10, 6))

# Line plot for rainfall over time
plt.plot(df['Date'], df['rainfall (mm)'], marker='o', linestyle='-', color='blue', label='Rainfall')

# Scatter plot for flooding events
plt.scatter(df['Date'], df['rainfall (mm)'] + 10, color='orange', marker='x', label='Flooding Event')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Rainfall (mm)')
plt.title('Rainfall Levels and Flooding Events Over Time')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Adding a legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python

# Calculate the total cost of floodings and the total number of people affected
total_cost = df['cost of damage (milion)'].sum()
total_people_affected = df['people affected by flooding'].sum()

# Create a bar chart
plt.figure(figsize=(8, 6))

# Bar plot
bars = plt.bar(['Total Cost', 'Total People Affected'], [total_cost, total_people_affected], color=['blue', 'orange'])

# Adding labels and title
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Total Cost and Total People Affected by Flooding')

# Adding values on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center', color='black', fontsize=10)

# Display the plot
plt.tight_layout()
plt.show()
```

- How much money did Slovakia spend into anty-flooding systems since 1990? 
According to the Slovak Water Management Authority, Slovakia has spent over €1.5 billion on anti-flooding systems since 1990. This includes the construction of levees, dams, and retention ponds, as well as the improvement of river channels.
- 1990-2000: €200 million
- 2001-2010: €500 million
- 2011-2020: €800 million

- The Slovak government has also received funding from the European Union for anti-flooding projects. In the period 2007-2013, the EU provided €200 million for anti-flooding measures in Slovakia.

- The Slovak government is committed to continuing to invest in anti-flooding systems. It is estimated that the country  will need to spend an additional €1 billion on anti-flooding measures in the next 10 years.
